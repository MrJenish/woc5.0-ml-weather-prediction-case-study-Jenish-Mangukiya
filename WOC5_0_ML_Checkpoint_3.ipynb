```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
cars = pd.read_csv("/content/CarPrice_Assignment.csv",encoding='cp1252')
```

```python
sns.kdeplot(cars.horsepower,shade=True)
```

```python
sns.distplot(cars.peakrpm)
```

```python
plt.figure(figsize=(8,8))
sns.boxplot(x=cars.symboling,y=cars.horsepower)
```

```python
sns.scatterplot(x='citympg',y='highwaympg',data=cars)
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(cars.corr(),square=True,lw=0.2)
```

```python
sns.swarmplot(x=cars.symboling, y=cars.compressionratio);
```

```python
plt.figure(figsize=(6,5))
sns.regplot(x='price',y='enginesize',data=cars,
            ci=None,
            scatter_kws={'s': 50, 'alpha': 0.8, 'color': 'lightgray'})
```

```python
sns.histplot(cars.enginesize)
```

```python
data1=cars[cars.symboling.isin([1,2,3])]
sns.histplot(x='stroke',data=data1,hue='symboling',element='step')
```

```python
data1=cars[cars.symboling.isin([0,-1,-2])]
sns.histplot(x='stroke',data=data1,hue='symboling',element='step')
```

```python
sns.barplot(cars.symboling,cars.enginesize)
```

```python
sns.barplot(cars.symboling,cars.peakrpm)
```
